```python

from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import tensorflow as tf

def plot_history(history):
  fig, (ax1, ax2) = plt.subplots(2,1, sharex=True, dpi=150)
  ax1.plot(history.history['loss'], label='training')
  ax1.plot(history.history['val_loss'], label='validation')
  ax1.set_ylabel('Loss')
  ax1.set_yscale('log')
  if history.history.__contains__('lr'):
    ax1b = ax1.twinx()
    ax1b.plot(history.history['lr'], 'g-', linewidth=1)
    ax1b.set_yscale('log')
    ax1b.set_ylabel('Learning Rate', color='g')
  ax1.legend()

  key = None
  for k in sorted(history.history.keys()):
    if 'accuracy' in k and not 'val_' in k:
      key = k
      break
  if key:
    ax2.plot(history.history[key], label='training')
    ax2.plot(history.history['val_'+key], label='validation')
    ax2.set_ylabel('Accuracy')
    ax2.set_xlabel('Epochs')
  plt.show()


def get_predicted_and_correct_labels(model, dataset):
  y_pred = [] # store predicted labels
  y_true = [] # store true labels

  # iterate over the dataset
  for image_batch, label_batch in dataset:   # use dataset.unbatch() with repeat
    # append true labels
    y_true.append(label_batch)
    # compute predictions
    preds = model.predict(image_batch)
    # append predicted labels
    y_pred.append(np.argmax(preds, axis=-1).astype('int32').flatten())

  # convert the true and predicted labels into tensors
  correct_labels = tf.concat([item for item in y_true], axis = 0).numpy()
  predicted_labels = tf.concat([item for item in y_pred], axis = 0).numpy()

  return predicted_labels, correct_labels#, images


def plot_confmat(Y, Y_true, class_names=None, normalize='true'):

  fig, ax = plt.subplots(figsize=(10, 5))
  disp = ConfusionMatrixDisplay.from_predictions(
      Y_true, Y,
      normalize=normalize,
      display_labels=class_names,
      ax=ax
  )
  acc = np.mean(Y_true==Y)
  class_averaged_accuracy = np.diag(disp.confusion_matrix).mean()
  titlestr = f"Accuracy: {acc*100:.2f}%"
  if normalize=='true':
    titlestr += f"\nClass-Averaged Accuracy: {class_averaged_accuracy*100:.2f}%"

  plt.xticks(rotation=45, ha='right')
  plt.title(titlestr)

  plt.grid(False)
  plt.show()
```
